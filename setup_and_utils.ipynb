```python
# Cài đặt môi trường và các thư viện cần thiết
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from PIL import Image
import time
import io
import zipfile
from google.colab import files
from google.colab.patches import cv2_imshow
from IPython.display import display, Javascript, Image as IPImage
import ipywidgets as widgets
import pickle

# Kiểm tra phiên bản TensorFlow và GPU
print(f"TensorFlow version: {tf.__version__}")
print("GPU Available: ", tf.config.list_physical_devices('GPU'))

# Tạo thư mục để lưu trữ các file chia sẻ giữa notebook
if not os.path.exists('shared'):
    os.makedirs('shared')

# Tạo các thư mục khác nếu cần
directories = ['models', 'dataset', 'images']
for directory in directories:
    if not os.path.exists(directory):
        os.makedirs(directory)

# Hàm tiện ích để chia sẻ biến giữa các notebook
def save_variable(variable, name):
    with open(f'shared/{name}.pkl', 'wb') as f:
        pickle.dump(variable, f)
    print(f"Saved {name} to shared/{name}.pkl")

def load_variable(name):
    try:
        with open(f'shared/{name}.pkl', 'rb') as f:
            variable = pickle.load(f)
        print(f"Loaded {name} from shared/{name}.pkl")
        return variable
    except:
        print(f"Could not load {name}")
        return None

# Hàm hiển thị ảnh
def display_image(image):
    if isinstance(image, np.ndarray):
        if len(image.shape) == 3 and image.shape[2] == 3:
            # Convert BGR to RGB for display
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        else:
            plt.imshow(image, cmap='gray')
    else:
        plt.imshow(image)
    plt.axis('off')
    plt.show()

# Kiểm tra cài đặt
print("Setup completed successfully!")
```
